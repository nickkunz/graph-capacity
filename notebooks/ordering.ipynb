## Dependencies

```python
## libraries
import os
import sys
import pandas as pd

## project root
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

## modules
from src.evaluators.config import FEAT_X, FEAT_Z, TARGET
from src.evaluators.config import load_data, load_models
from src.evaluators.ordering import eval_order
```

## Initalization

```python
## setup
data = load_data()
models = load_models()
```

## Training

```python
## structural ordering results
results_list = list()
for name, model in models.items():
    print(f"training {name}...")
    metrics = eval_order(
        data = data,
        feat_x = FEAT_X,
        feat_z = FEAT_Z,
        estimator_c = model.estimator_c,
        estimator_r = model.estimator_r,
        target = TARGET
    )
    
    ## append model name and store
    metrics["model"] = name
    results_list.append(metrics)
```

## Post-Processing

```python
## convert to dataframe
results_data = pd.DataFrame(results_list)
feat = ["model"] + [c for c in results_data.columns if c != "model"]
results_data = results_data[feat]
```

## Results

```python
## preview results
results_data.mean(numeric_only = True)
```
